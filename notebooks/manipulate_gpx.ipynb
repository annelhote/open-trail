```python
# https://pypi.org/project/gpxpy/
import gpxpy
import requests
```

```python
def merge_points(urls):
  points = []
  for url in urls:
    url_requested = requests.get(url)
    gpx = gpxpy.parse(url_requested.text)
    for track in gpx.tracks:
      for segment in track.segments:
        for point in segment.points:
          points.append(point)
  gpx = gpxpy.gpx.GPX()
  # Create first track in our GPX:
  gpx_track = gpxpy.gpx.GPXTrack()
  gpx.tracks.append(gpx_track)
  # Create first segment in our GPX track:
  gpx_segment = gpxpy.gpx.GPXTrackSegment()
  gpx_track.segments.append(gpx_segment)
  # Create points:
  gpx_segment.points = points
  with open("merged_points.gpx", "w") as file:
    file.write(gpx.to_xml())
```

```python
def merge_waypoints(from_file, to_files):
  from_file = open(from_file, "r")
  from_content = from_file.read()
  from_gpx = gpxpy.parse(from_content)
  waypoints = from_gpx.waypoints
  from_file.close()
  for file_name in to_files:
    file_content = open(file_name, "r")
    to_gpx = gpxpy.parse(file_content.read())
    for waypoint in to_gpx.waypoints:
      waypoints.append(waypoint)
    file_content.close()
  # TODO : deduplicate waypoints
  # Create waypoints:
  from_gpx.waypoints = waypoints
  with open("merged_waypoints.gpx", "w") as file:
    file.write(from_gpx.to_xml())
```
